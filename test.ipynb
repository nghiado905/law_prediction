```python
import os
import requests
from bs4 import BeautifulSoup
import pdfplumber
from docx import Document

# Thư mục lưu trữ các văn bản luật
SAVE_DIR = 'test'
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

# Danh sách các nguồn dữ liệu
URLS = [
    'https://vanban.chinhphu.vn/he-thong-van-ban?classid=1&mode=1&typegroupid=1'
]

def download_file(url, save_path):
    try:
        response = requests.get(url)
        response.raise_for_status()
        with open(save_path, 'wb') as file:
            file.write(response.content)
        print(f"Đã tải xuống {url} và lưu tại {save_path}")
    except requests.RequestException as e:
        print(f"Không thể tải xuống {url}: {e}")

def process_pdf(file_path):
    with pdfplumber.open(file_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    with open(file_path.replace('.pdf', '.txt'), 'w', encoding='utf-8') as text_file:
        text_file.write(text)
    print(f"Đã trích xuất văn bản từ {file_path}")

def process_docx(file_path):
    doc = Document(file_path)
    text = ''
    for para in doc.paragraphs:
        text += para.text + '\n'
    with open(file_path.replace('.docx', '.txt'), 'w', encoding='utf-8') as text_file:
        text_file.write(text)
    print(f"Đã trích xuất văn bản từ {file_path}")

def crawl_data():
    for url in URLS:
        try:
            response = requests.get(url)
            response.raise_for_status()
            soup = BeautifulSoup(response.text, 'html.parser')
            
            links = soup.find_all('a', href=True)
            for link in links:
                href = link['href']
                if href.endswith('.pdf'):
                    download_file(href, os.path.join(SAVE_DIR, os.path.basename(href)))
                elif href.endswith('.docx'):
                    download_file(href, os.path.join(SAVE_DIR, os.path.basename(href)))
        
        except requests.RequestException as e:
            print(f"Không thể truy cập {url}: {e}")

def main():
    crawl_data()

    # Xử lý các tập tin đã tải xuống
    for file_name in os.listdir(SAVE_DIR):
        file_path = os.path.join(SAVE_DIR, file_name)
        if file_name.endswith('.pdf'):
            process_pdf(file_path)
        elif file_name.endswith('.docx'):
            process_docx(file_path)

if __name__ == "__main__":
    main()
```

```python
import os
import requests
from bs4 import BeautifulSoup

# Thư mục lưu trữ các tài liệu
SAVE_DIR = 'attachments'
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

# Hàm tải tệp DOCX từ một trang
def download_docx_from_page():
    base_url = f'https://vanban.chinhphu.vn/he-thong-van-ban?classid=1&mode=1&typegroupid=1'
    
    # Gửi yêu cầu HTTP và lấy nội dung HTML của trang
    response = requests.get(base_url)
    response.raise_for_status()  # Kiểm tra lỗi HTTP
    
    # Phân tích HTML
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Tìm tất cả các liên kết tài liệu đính kèm
    attachment_links = soup.find_all('a', href=True)
    
    # Lọc các liên kết tài liệu đính kèm (DOCX)
    docx_links = [link['href'] for link in attachment_links if link.get_text(strip=True) == 'Tài liệu đính kèm']
    
    
    # Lọc và tải các tài liệu DOCX
    for href in docx_links:
        file_url = href if href.startswith('http') else f"https://vanban.chinhphu.vn{href}"
        file_name = os.path.basename(file_url)
        save_path = os.path.join(SAVE_DIR, file_name)
        
        # In ra đường dẫn để kiểm tra
        print(f"Tải xuống: {file_url}")
        
        if not os.path.exists(save_path):
            try:
                # Tải file
                file_response = requests.get(file_url)
                file_response.raise_for_status()  # Kiểm tra lỗi HTTP
                with open(save_path, 'wb') as file:
                    file.write(file_response.content)
                print(f"Đã tải xuống {file_url} và lưu tại {save_path}")
            except requests.RequestException as e:
                print(f"Lỗi khi tải {file_url}: {e}")



download_docx_from_page()
```

```python
import os
import re
import requests
from bs4 import BeautifulSoup
import zipfile

# Thư mục lưu trữ các tài liệu
SAVE_DIR = 'hien_phap'
if not os.path.exists(SAVE_DIR):
    os.makedirs(SAVE_DIR)

# Hàm tải tệp từ một trang
def download_files_from_page():
    base_url = 'https://vbpl.vn/TW/Pages/vanban.aspx?idLoaiVanBan=15&dvid=13'
    
    # Gửi yêu cầu HTTP và lấy nội dung HTML của trang
    response = requests.get(base_url)
    response.raise_for_status()  # Kiểm tra lỗi HTTP
    
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Lấy tất cả các liên kết đính kèm
    attachment_links = soup.find_all('a', href=True)
    
    # Trích xuất các liên kết tệp từ mã JavaScript
    file_links = []
    for link in attachment_links:
        href = link['href']
        if 'javascript:downloadfile' in href:
            # Sử dụng regex để trích xuất URL tệp
            match = re.search(r"downloadfile\('([^']+)',\s*'([^']+)'\)", href)
            if match:
                file_name = match.group(1)
                file_url = f"https://vbpl.vn{match.group(2)}"
                file_links.append((file_url, file_name))
    
    print(f"Tìm thấy các liên kết trên trang: {file_links}")

    # Tải xuống và giải nén các file ZIP
    for file_url, file_name in file_links:
        save_path = os.path.join(SAVE_DIR, file_name)
        
        # In ra đường dẫn để kiểm tra
        print(f"Tải xuống: {file_url}")
        
        if not os.path.exists(save_path):
            try:
                # Tải file
                file_response = requests.get(file_url)
                file_response.raise_for_status()  # Kiểm tra lỗi HTTP
                with open(save_path, 'wb') as file:
                    file.write(file_response.content)
                print(f"Đã tải xuống {file_url} và lưu tại {save_path}")
            except requests.RequestException as e:
                print(f"Lỗi khi tải {file_url}: {e}")
                
            # Giải nén nếu là tệp ZIP
            if file_name.lower().endswith('.zip'):
                try:
                    with zipfile.ZipFile(save_path, 'r') as zip_ref:
                        zip_ref.extractall(SAVE_DIR)
                        print(f"Đã giải nén {save_path} vào thư mục {SAVE_DIR}")
                except zipfile.BadZipFile:
                    print(f"Tệp {save_path} không phải là ZIP hợp lệ.")
                finally:
                    # Xóa tệp ZIP sau khi giải nén
                    os.remove(save_path)
                    print(f"Đã xóa tệp ZIP {save_path}")

# Gọi hàm để tải file và giải nén nếu cần
download_files_from_page()
```
